## Логічі операції із зображеннями

```python
## Завантаження пакетів
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Генерация  першого изображения 
rows_num = 200 ## кількість рядків
clms_num = 200 ## кількість колонок
im_1 = np.full ((rows_num, clms_num, 3), (255, 255, 255.) , dtype=np.int32)
for j in  range (0, clms_num-1, 1):
    for i in  range (0, 24, 1):
            im_1 [i, j, : ] = [0, 0, 0]
            im_1 [i+50, j, : ] = [0, 0, 0]
            im_1 [i+100, j, : ] = [0, 0, 0]
            im_1 [i+150, j, : ] = [0, 0, 0]

plt.title('IM 1')
plt.imshow(im_1)
plt.show()
```

```python
## Генерация  другого изображения 
im_2 = np.full ((rows_num, clms_num, 3), (255, 255, 255.) , dtype=np.int32)
for i in  range (0, rows_num-1, 1):
    for j in  range (0, 24, 1):
            im_2 [i, j, : ] = [0, 0, 0]
            im_2 [i, j+50, : ] = [0, 0, 0]
            im_2 [i, j+100, : ] = [0, 0, 0]
            im_2 [i, j+150, : ] = [0, 0, 0]
            
plt.title('IM 2')
plt.imshow(im_2)
plt.show()
```

####  1. Одномістна логічна операція NOT (IM_1) 

```python
im_not = np.full ((rows_num, clms_num, 3), (255, 255, 255.) , dtype=np.int32)
for i in  range (0, rows_num-1, 1):
    for j in  range (0, clms_num-1, 1):
            im_not [i, j, : ] = 255 - im_1 [i, j, : ]
            
plt.title('IM NOT')
plt.imshow(im_not)
plt.show()
```

####  2. Двомістна логічна операція AND (IM_1 & IM_2)

```python
im_and = np.full ((rows_num, clms_num, 3), (0, 0, 0) , dtype=np.int16)

for i in  range (0, rows_num-1, 1):
    for j in  range (0, clms_num-1, 1):
        if im_1[i,j,0]==255 and im_2[i,j,0]==255 :
            im_and [i, j, : ] = [255, 255, 255]
            
plt.title('IM AND')
plt.imshow(im_and)
plt.show()
```

####  3. Двомістна логічна операція OR (IM_1 V IM_2)

```python
im_or = np.full ((rows_num, clms_num, 3),  (0, 0, 0) , dtype=np.int16)

for i in  range (0, rows_num-1, 1):
    for j in  range (0, clms_num-1, 1):
            if im_1[i,j,0]==255 or im_2[i,j,0]==255 :
                im_or [i, j, : ] = [255, 255, 255]
            
plt.title('IM OR')
plt.imshow(im_or)
plt.show()
```

####  4. Двомістна логічна операція XOR (IM_1 xor IM_2)

```python
im_xor = np.full ((rows_num, clms_num, 3),  (0, 0, 0) , dtype=np.int16)

for i in  range (0, rows_num-1, 1):
    for j in  range (0, clms_num-1, 1):
            if (im_1[i,j,0]==255 and im_2[i,j,0]==255) or (im_1[i,j,0]==0 and im_2[i,j,0]==0) :
                im_xor [i, j, : ] = [255, 255, 255]
            
plt.title('IM XOR')
plt.imshow(im_xor)
plt.show()
```
