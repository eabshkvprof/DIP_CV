## Створення / завантаження зображень

```python
## Завантаження пакетів
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io
plt.rcParams['font.size'] = 10
```

#### 1. Створення ахроматичного зображення (біле / чорне)

```python
rows_num = 200 ## кількість рядків
clms_num = 512 ## кількість стовпчиків

# test_im = np.full ((rows_num, clms_num, 3), (0, 0, 0) , dtype=np.uint8) # чорне
# test_im = np.full ((rows_num, clms_num, 3), (250, 250, 250) , dtype=np.uint8) # біле
test_im = np.full ((rows_num, clms_num, 3), (1., 1., 1.) , dtype=np.float32) # біле (з плаваючою комою)
print ('IMAGE SHAPE', test_im.shape, 'IMAGE SIZE', test_im.size)
```

```python
## Вивід оррігінального зображення на екран
fig, ax = plt.subplots(figsize=(10, 6))
plt.title('ORIGINAL IMAGE')
plt.imshow(test_im)
plt.show()
```

#### 2. Створення ахроматичного зображення (градієнт: від чорного до білого)

```python
test_im_ = np.full ((rows_num, clms_num, 3), (0, 0, 0) , dtype=np.uint8)
for i in  range (rows_num):
    for j in  range (clms_num):
        test_im_ [i, j, : ] = [j/2, j/2, j/2]
        
```

```python
## Вивід оррігінального зображення на екран
fig, ax = plt.subplots(figsize=(10, 6))
plt.title('ORIGINAL IMAGE')
plt.imshow(test_im_)
plt.show()
```

#### 3. Завантаження кольорового зображення

```python
## Завантаження файлу зображення
path = './IMAGES/' 
filename = 'Lenna.png' 
test_im = io.imread(path + filename)
## Визначення стркутури та розміру зображення
print ('IMAGE SHAPE', test_im.shape, 'IMAGE SIZE', test_im.size)
rows_num = test_im.shape[0] ## кількість рядків
clms_num = test_im.shape[1] ## кількість колонок
pix_num = rows_num*clms_num ## кількість пікселів
bins = 256 ## кількість рівнів яскравості
print ('ROWS NUMBER', rows_num, 'CLMS NUMBER', clms_num, 'PIX NUMBER', pix_num, 'Bins',bins)
```

```python
## Вивід оррігінального зображення на екран
fig, ax = plt.subplots(figsize=(5, 3))
plt.title('ORIGINAL IMAGE')
plt.imshow(test_im)
plt.show()
```

#### 4. Перетворення кольорового до ахроматичного (сірого)

```python
### ФОРМУВАННЯ НАПІВТОНОВОГО ЗОБРАЖЕННЯ
gray_im = np.zeros ( (rows_num, clms_num, 3), dtype=np.uint8)

for i in  range (rows_num):
    for j in  range (clms_num):
        # Gray image
        gray_im [i, j, :] = 0.299*test_im [i, j, 0]+0.587*test_im [i, j, 1]+0.114*test_im [ i, j, 2]        
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
plt.title('GRAY IMAGE')
plt.imshow(gray_im)
plt.show()
```
